```python
# Extract the gender column
gender_data = df['gender']
```

```python
df.columns = df.columns.str.strip()
```

```python
import matplotlib.pyplot as plt
```

```python
gender_counts = gender_data.value_counts()
```

```python
 # Create a bar chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.xlabel('gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Create a pie chart
plt.figure(figsize=(8, 8))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=135)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```
